### Abstract

This dataset is a multilabel classification problem. The goal is to predict the presence of amphibians species near the water reservoirs based on features obtained from GIS systems and satellite images. <br>
Dataset: https://archive.ics.uci.edu/ml/datasets/Amphibians

### Data Set Information

The data used was derived from GIS and satellite information, as well as from information gathered from the natural inventories that were prepared for the environmental impact assessment (EIA) reports for two planned road projects (Road A and Road B) in Poland. These reports were mostly used to gather information on the size of the amphibian population in each of the 189 occurrence sites.
Road A project concerned part of the planned A1 motorway section in Pyrzowice; the section is located along the northern border of the Silesian Voivodship and is about 75 km long. The field research involved a strip of land with a width of 500 m on both sides of the proposed project area. The field inventory was carried out in 2010 [1] and 2011 [2,3]. The results of these inventories were complemented by Marek SoÅ‚tysiak observations, which were conducted between 2014 and 2016 [4]. Finally, the first project included 80 amphibian breeding sites.
Road B inventory was prepared in the vicinity of two variants of the planned Beskidy Integration Way on the Bielsko BiaÅ‚a-Wadowice-GÅ‚ogoczÃ³w section of the S52 motorway. The length of this section of road is approximately 60 km. During the inventory, which was taken in 2010, 125 real and potential amphibian occurrence sites were described in [5]. The methodology of the herpetological inventory included map analysis, literature and archive data analysis, and, then, field observations. As in the first case, the inventory was made in the spring time and consisted of the observation of the occurrence of amphibians in water reservoirs. The research area included a 500 m wide belt for each of the considered variants of the planned road. In order to conduct the final experiments, 109 amphibian occurrence sites were taken into account.

## Decision Tree Classifier

### Predicting Green Frogs

```python
target = 'Green frogs'
predictors = df.columns[df.columns != target]
X = df[predictors].values
y = df[target].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # Make the test size 20%
```

###### Baseline Accuracy

The first thing to do is to find the baseline accuracy. Our baseline accuracy is computed by creating an array of all 0's (Not present) and comparing it against our test data.

```python
print("Our baseline accuracy is {:.2f}%".format((np.zeros(y_test.shape[0], dtype=int) == y_test).mean() * 100))
```

###### Model 1 (Default Hyperparameters)

For the first model, i will not set any hyperparameters

```python
clf = DecisionTreeClassifier(random_state=0) # Create the decision tree
clf.fit(X_train, y_train) # Train the model
```

Make predictions

```python
y_predict = clf.predict(X_test)
print('The Accuracy of the first model is {:.2f}%'.format((y_test == y_predict).mean()*100))
```

The first models accuracy was 76.32%. This is a huge improvement from our baseline accuracy. Next, i will be looking at the confusion matrix and comparing the precision, recall, and f-score.

```python
def print_conf_mtx(y_true, y_pred, classes=None):
    """ Print a confusion matrix (two classes only). """
    
    if not classes:
        classes = ['False', 'True']
   	 
    # formatting
    max_class_len = max([len(s) for s in classes])
    m = max(max_class_len, len('predicted')//2 + 1)
    n = max(len('actual')+1, max_class_len)
    left   	= '{:<10s}'.replace('10',str(n))
    right  	= '{:>10s}'.replace('10',str(m))
    big_center = '{:^20s}'.replace('20',str(m*2))
    
    cm = confusion_matrix(y_test, y_pred)
    print((left+big_center).format('', 'predicted'))
    print((left+right+right).format('actual', classes[0], classes[1]))
    print((left+right+right).format(classes[0], str(cm[0,0]), str(cm[0,1])))
    print((left+right+right).format(classes[1], str(cm[1,0]), str(cm[1,1])))

print_conf_mtx(y_test, y_predict)
```

The confusion matrix shows us that we correctly predicted 29 out of 38 test observations. Next, i will compute the precision, recall, and f-scores. 

```python
from sklearn.metrics import precision_recall_fscore_support
def metrics(y_test, y_predict):
    scores = precision_recall_fscore_support(y_test, y_predict, average='binary')
    print('Precision: {}%'.format(round(scores[0],4)*100))
    print('Recall: {}%'.format(round(scores[1],4)*100))
    print('F-score: {}%'.format(round(scores[2],4)*100))

metrics(y_test, y_predict)
```

Our overall precision, recall, and f-score are all pretty good. Next, i will try to improve the next model by using feature importance

```python
#importance = np.sort(clf.feature_importances_ * 100)[::-1]
importance = clf.feature_importances_ * 100

ax = pd.DataFrame(importance, index=predictors, columns=['Importance']).plot.bar()
plt.title('Importance of Features')
plt.xticks(rotation=45, horizontalalignment='right')
plt.ylabel('Percentage')
# Add the numbers to the graph
for i, col in enumerate(ax.patches):
    plt.text(col.get_x()-0.05, col.get_height(), round(importance[i],2), fontsize=font_size-5);
```

From the graph above, NR, MR, CR, Common toad, Tree frog, and Great crested newt did not contribute to the decision tree. 

<b> Model 2 </b>

Next, i will create a new model with the features with the greatest importance. 

```python
important_features = ['SR','TR','VR','SUR1','SUR2',
                      'SUR3','UR','OR','RR','BR','Brown frogs',
                      'Fire-bellied toad','Common newt']

target = 'Green frogs'
X = df[important_features].values
y = df[target].astype(int)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

clf = DecisionTreeClassifier(random_state=0) 
clf.fit(X_train, y_train) # Train the model

# Make predictions
y_predict = clf.predict(X_test)
print('The Accuracy of the second model is {:.2f}%'.format((y_test == y_predict).mean()*100))
```

Our accuracy went down from the first model. Next, i will look at the confusion matrix and look at some metrics again.

```python
print_conf_mtx(y_test, y_predict)
```

The only difference between the first model's confusion matrix and the second one is there is one value was removed from the True Negative and added to the False Positive. Because the training set is so small, it added a large change in accuracy.

```python
metrics(y_test, y_predict)
```

Because the False Positive and False Negative are the same values, it means we will have the same score for our precision, recall, and f-score. Next, i will create one more model using the important features from the previous model and use grid search to find the optimal hyperparamters. 

<b> Model 3 </b>

Using grid search and feature importance from the last model

```python
params = {'criterion':['gini','entropy'],
          'max_depth':range(1,20),
          'min_samples_split':range(1,20),
          'min_samples_leaf':range(1,20),
          'splitter':['best','random']
}
grid = GridSearchCV(DecisionTreeClassifier(random_state=0), param_grid=params, cv=5, n_jobs=-1, scoring='accuracy')
grid.fit(X_train, y_train)
```

These are the best parameters determined by the grid search

```python
metrics(y_test, y_predict)
```

The precision and f-score dropped compared to the last two models but the precision score increased. 

Next, i will try to find the optimal threshold for all of the metrics.

```python
# This function takes the y_test, x_test, and model and computes the greatest precision, recall, f-score, and accuracy
# and return all the values of them from a threshold of 0 to 100
def get_best_threshold(y_test, X_test, model):
    
    best_precision = {0:0} # key (threshold) -> 0, value (score) -> 0
    best_recall = {0:0}
    best_fscore = {0:0}
    best_accuracy = {0:0}
    scores_list = []
    
    y_proba = model.predict_proba(X_test)
    for i in range(0,101):
        y_predict = y_proba[:,1] > i/100
        scores = precision_recall_fscore_support(y_test, y_predict, average='binary', zero_division=0)
        scores_list.append(scores[:3])
        
        # Best precision, recall, and f-score
        if scores[0] > list(best_precision.values())[0]:
            best_precision.clear()
            best_precision[i] = scores[0]
        if scores[1] > list(best_recall.values())[0]:
            best_recall.clear()
            best_recall[i] = scores[1]
        if scores[2] > list(best_fscore.values())[0]:
            best_fscore.clear()
            best_fscore[i] = scores[2]    
        
        # Compute the accuracy
        accuracy = (y_test == y_predict).mean()
        if accuracy > list(best_accuracy.values())[0]:
            best_accuracy.clear()
            best_accuracy[i] = accuracy
            
    return best_precision, best_recall, best_fscore, best_accuracy, scores_list

best_precision, best_recall, best_fscore, best_accuracy, scores_list = get_best_threshold(y_test, X_test, grid)
```

```python
print('Best Accuracy {:.2f}% with the threshold 0.{}'.format(list(best_accuracy.values())[0] * 100, list(best_accuracy.keys())[0]))
print('Best Precision {:.2f}% with the threshold 0.{}'.format(list(best_precision.values())[0] * 100, list(best_precision.keys())[0]))
print('Best Recall {:.2f}% with the threshold 0.{}'.format(list(best_recall.values())[0] * 100, list(best_recall.keys())[0]))
print('Best F-Score {:.2f}% with the threshold 0.{}'.format(list(best_fscore.values())[0] * 100, list(best_fscore.keys())[0]))
```

An accuracy of only 71% is not that great compared to the first model with no set hyperparameters. 

The precision of 89% is a great score. With the best recall and f-score being at the threshold of 0 not being that helpful, I will next plot the precision and recall curve as well as plotting them individually. 

```python
# Get the precision and recal scores
prec = []
rec = []
for i in range(0,101):
    prec.append(scores_list[i][0])
    rec.append(scores_list[i][1])

fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(range(0,101), prec, label='Precision', marker='.')
ax1.plot(range(0,101), rec, label='Recall', marker='.')
ax1.set_title('Precision vs Recall')
ax1.set_xlabel('Threshold')
ax1.set_ylabel('Score')
ax1.legend(loc=3)

plot_precision_recall_curve(clf, X_test, y_test, ax=ax2)
ax2.set_title('2-class Precision-Recall curve')

plt.tight_layout();
```
